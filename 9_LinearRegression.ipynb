## Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
X = np.random.rand(100, 1)
y = 3 + 4 * X + np.random.rand(100, 1)
```

```python
plt.scatter(X, y)
plt.show()
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
plt.figure()
plt.scatter(X, y)
plt.plot(X, lin_reg.predict(X), color = 'red', alpha=0.7)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression')
plt.show()
```

```python
# y = 3 + 4 * X
a1 = lin_reg.coef_[0][0]
a0 = lin_reg.intercept_[0]
a0, a1
```

```python
for i in range(100):
  y_ = a0 + a1 * X
  plt.plot(X, y_, color='green', alpha=0.7)
```

## Linear Regression on Diabete Dataset

```python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import numpy as np
```

```python
diabetes = load_diabetes()
# or
diabetes_X, diabetes_y = load_diabetes(return_X_y=True)

diabetes_X = diabetes_X[:, np.newaxis, 2]    # only bmi
diabetes_X[:10]
```

```python
diabetes_X_train = diabetes_X[:-20]
diabetes_X_test = diabetes_X[-20:]
diabetes_y_train = diabetes_y[:-20]
diabetes_y_test = diabetes_y[-20:]
diabetes_X_train.shape, diabetes_X_test.shape, diabetes_y_train.shape, diabetes_y_test.shape
```

```python
lig_reg = LinearRegression()
lig_reg.fit(diabetes_X_train, diabetes_y_train)
diabetes_y_pred = lig_reg.predict(diabetes_X_test)
```

```python
mean_squared_error(diabetes_y_pred, diabetes_y_test), r2_score(diabetes_y_pred, diabetes_y_test)
```

```python
plt.scatter(diabetes_X_test, diabetes_y_test, color = 'black')
plt.plot(diabetes_X_test, diabetes_y_pred, color = 'green')
plt.show()
```
